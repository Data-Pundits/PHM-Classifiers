```python
fault_list = phm_df['fault_class'].values.tolist()
fault_list = list(set(fault_list))
features_list = list(phm_df.columns)
features_list = [x for x in features_list if x not in ['pdmp', 'pin', 'po']]

def Query1Viz(faultSelection, xFeatureSelection, yFeatureSelection):
    filtered_phm_df = phm_df[phm_df['fault_class'] == faultSelection]
    xList = filtered_phm_df[xFeatureSelection].values.tolist()
    yList = filtered_phm_df[yFeatureSelection].values.tolist()

    figure(figsize=(12, 6), dpi=80)
    plt.scatter(xList, yList, c=filtered_phm_df['individual'])

    plt.title('For Fault Class ' + str(faultSelection))
    plt.xlabel(xFeatureSelection)
    plt.ylabel(yFeatureSelection)

    plt.show()

widgets.interact(Query1Viz, faultSelection=fault_list, xFeatureSelection=features_list, yFeatureSelection=features_list)
```
