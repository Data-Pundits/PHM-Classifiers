```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
phm_df = pd.read_parquet('../feature_extracts/', engine='fastparquet')
```

```python
f1_phm_df = phm_df[phm_df['fault_class'] == 1]
f1_phm_df
```

```python
x_list = np.arange(1, 572)
y_pdmp_list = f1_phm_df['pdmp'].values[0:3]
y_pin_list = f1_phm_df['pin'].values[0:3]
y_po_list = f1_phm_df['po'].values[0:3]
print(len(y_pdmp_list))
print(y_pdmp_list)
```

```python
plt.plot(x_list, y_pdmp_list)
plt.plot(x_list, y_pin_list)
plt.plot(x_list, y_po_list)

plt.title('For Fault 1:')
plt.xlabel('Time')
plt.ylabel('Pressure')
plt.show()
```
